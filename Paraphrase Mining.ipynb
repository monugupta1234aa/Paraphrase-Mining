```python
s1 = "the is on the mat "
s2 = "a cat is sitting on the mat "
emb1= model.encode(s1)
emb2= model.encode(s2)
sim_score = util.cos_sim(emb1,emb2)
sim_score
```

```python
corpus1= ["the stock ,market saw a big drop",
          "shares declined sharply due to economic news.",
          "money gets money in stock",
          "It may rain today"]
corpus_emb = model.encode(corpus1)
```

```python
cos_similarity = util.pytorch_cos_sim(corpus_emb,corpus_emb)
avg_sim = cos_similarity.mean(dim=1)
outlier_index = avg_sim.argmin().item()
outlier_index
print(f"outlier sentence : {corpus1[outlier_index]} (AVG Similarity : {avg_sim[outlier_index]})")
```
